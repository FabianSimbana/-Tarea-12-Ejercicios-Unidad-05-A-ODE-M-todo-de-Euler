## 1.Use el m√©todo de Euler para aproximar las soluciones para cada uno de los siguientes problemas de valor inicial.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definici√≥n de la funci√≥n derivada
def f(t, y):
    return t * np.exp(3*t) - 2*y

# Par√°metros iniciales
h = 0.5
t0 = 0
y0 = 0
tf = 1
n_steps = int((tf - t0) / h)

# Inicializaci√≥n de listas para almacenar los resultados
t_values = [t0]
y_values = [y0]

# M√©todo de Euler
for i in range(n_steps):
    t_n = t_values[-1]
    y_n = y_values[-1]
    
    y_next = y_n + h * f(t_n, y_n)
    t_next = t_n + h
    
    t_values.append(t_next)
    y_values.append(y_next)

print(t_values, y_values)

# Graficar los resultados
plt.plot(t_values, y_values, 'o-', label='Soluci√≥n Euler')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Ejercicio 1.a')
plt.grid(True)
plt.legend()
plt.show()
```

## 3.	Utilice el m√©todo de Euler para aproximar las soluciones para cada uno de los siguientes problemas de valor inicial.

```python
# Definici√≥n de la funci√≥n derivada
def f(t, y):
    return (y/t) - (y/t)**2

# Par√°metros iniciales
h = 0.1
t0 = 1
y0 = 1
tf = 2
n_steps = int((tf - t0) / h)

# Inicializaci√≥n de listas para almacenar los resultados
t_values = [t0]
y_values = [y0]

# M√©todo de Euler Impl√≠cito
for i in range(n_steps):
    t_n = t_values[-1]
    y_n = y_values[-1]
    
    # Estimaci√≥n inicial usando Euler expl√≠cito
    y_guess = y_n + h * f(t_n, y_n)
    
    # Correcci√≥n iterativa para aproximar el valor de y_n+1
    y_next = y_guess  # Iniciar con la estimaci√≥n de Euler expl√≠cito
    for _ in range(10):  # N√∫mero de iteraciones para convergencia
        y_next = y_n + h * f(t_n + h, y_next)
    
    t_next = t_n + h
    
    t_values.append(t_next)
    y_values.append(y_next)

# Graficar los resultados
plt.plot(t_values, y_values, 'o-', label='Soluci√≥n Euler Impl√≠cito')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Soluci√≥n usando M√©todo de Euler Impl√≠cito')
plt.grid(True)
plt.legend()
plt.show()
```
